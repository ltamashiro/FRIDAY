```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

%matplotlib inline
```

```python
# Creating DataFrame
df = pd.read_excel('dataset.xlsx',sheet_name='data_set')
```

```python
df.head()
```

```python
df.info()
```

```python
# Cheking Null Values
heat_missing = sns.heatmap(df.isnull(), cbar=False)
fig = heat_missing.get_figure()
fig.savefig("heat_missing.png",bbox_inches="tight") 
```

```python
# Handling missing values
values = {'comprehensive_product': 'TP', 'number_of_payment_faults': 0,'risk_predictor_zip_code':0}
df.fillna(value=values,inplace=True)
```

```python
# Cheking after data cleansing
sns.heatmap(df.isnull(), cbar=False)
```

```python
df['risk_predictor_zip_code'] = df['risk_predictor_zip_code'].astype(int)
df['number_of_payment_faults'] = df['number_of_payment_faults'].astype(int)
```

```python
df.head()
```

```python
df.to_csv('dataset_3.csv', sep=',', encoding='utf-8',index=False)
```
